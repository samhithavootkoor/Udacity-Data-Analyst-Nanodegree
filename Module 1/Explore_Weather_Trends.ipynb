```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
global_data=pd.read_csv('results.csv')
global_data
plt.plot(global_data.year,global_data.moving_avg,label='Global')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Temperature in Celsius')
plt.title('Global trends')
plt.show()
```

```python
city_data=pd.read_csv('citydata.csv')
city_data
plt.plot(global_data.year,global_data.moving_avg,label='Global')
plt.plot(city_data.year,city_data.moving_avg,label='Local')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Temperature in Celsius')
plt.title('Global vs Local trends')
plt.show()
```

```python
plt.plot(city_data.year,city_data.moving_avg,label='Local city trends')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Temperature in Celsius')
plt.title('City trends')
plt.show()
```
